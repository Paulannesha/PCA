# Dimensionality Reduction with PCA

```python
std = StandardScaler()
dt_normal=std.fit_transform(dt)
dt_normal
```

```python
pca2=PCA(n_components=None)
data = pca2.fit_transform(dt_normal)
pca_components=pd.DataFrame(data,columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11','pc12','pc13','pc14'])
pca_components
```

```python
pca_components.corr()
```

```python
plt.scatter(pca_components.pc1,pca_components.pc2)
```

```python
var = pca2.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color="red")
plt.title('Scree plot')
plt.xlabel("PC")
plt.ylabel("Cumulative Variance Ratio")
plt.grid()
plt.show()
```

# Clustering with Original Data:

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1, 20):
    
    kmeans = KMeans(n_clusters=i,random_state=0 )
    kmeans.fit(dt_normal)
    
    wcss.append(kmeans.inertia_)
print(wcss)    
plt.plot(range(1, 20), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(5, random_state=42)
clusters_new.fit(dt_normal)
```

```python
clusters_new.labels_
```

```python
set(clusters_new.labels_)
```

```python
dt['kmeans'] = clusters_new.labels_
dt
```

```python
dt1=dt.sort_values(by = "kmeans")
dt1
```

```python
dt1.groupby('kmeans').agg(['mean']).reset_index()
```

```python
pca2=PCA(n_components=None)
data = pca2.fit_transform(dt1)
pca_components2=pd.DataFrame(data,columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11','pc12','pc13','pc14','pc15'])
pca_components2
```

```python
finalDt = pd.concat([pd.DataFrame(pca_components2.iloc[:,0:2],columns=['pc1','pc2']), dt1[['kmeans']]], axis = 1)
sns.scatterplot(data=finalDt,x='pc1',y='pc2',hue='kmeans')
```
